#### specify that y will be a function of x 

```python
y = Function('y')(x) 
```

#### specify that the function $y(x) = e^x$

```python
# earlier, we declared y to be a function of x: y = Function('y')(x)
# now, let's specify y(x) to be the exponential function $e^x$ 

y = exp(x)   # specify y(x) 
print(y)     # print y(x) in plain form
y            #print y(x) in typeset form
```

```python
y.diff(x)
```

```python
result = y.diff(x) # ask for the first derivative of y with respect to x 
result                    # print result in typeset form
```

#### check if derivative equals function, as it must for the exponential

```python
# does result = y numerically? 
y.diff(x).equals(y) 
```
