# Eigenvalues and Eigenvectors

## find eigenvalues of a matrix

```python
M0 = Matrix([[2, -1], [-1, 2]])
M0
```

```python
M0 = Matrix([[2, -1], [-1, 2]])
M0.eigenvals()
```

This says 
- the first eigenvalue is 1, with a multiplicity of 1 (unique)
- the second eigenvalue is 3, with a multiplicity of 1 (unique)

```python
M = Matrix([[3, -2,  4, -2], [5,  3, -3, -2], [5, -2,  2, -2], [5, -2, -3,  3]])
M
M.eigenvals()
```

This says 
- the first eigenvalue is -2, with a multiplicity of 1 (unique)
- the second eigenvalue is 3, with a multiplicity of 1 (unique)
- the third and fourth eigenvalue are 5, thus 5 has a multiplicity of 2

# Alternatives 
Solving a matrix equation is equivalent to solving a system of linear equations, so if you prefer you can Solve a System of Equations Algebraically or use Row Reduction. 

## Solving a System of Equations Algebraically

```python
from sympy.abc import x, y, z
```

```python
# equation: y + 4*z = 0 
# we know the solution to this is y = -4z 
equation_1 = y + 4*z 
equation_1
```

```python
# equation: x + y -2z = 0
equation_2 = x + y - 2 *z 
equation_2
```

```python
solve([equation_1, equation_2], [x, y], dict=True)
```

### Row reduction for solving matrices 
This is an alternative to finding the matrix inverse

```python
row_1 = [1, -1,  2, 10] # 1x -1y +2z = 10
row_2 = [3,  0, -2, 11] # 3x -0y -2z = 11
row_3 = [2, 4,  -6, -6] # 2x +4y -6z = -6

M = Matrix([row_1, row_2, row_3])
M
```

```python
M.rref()[0]
```

the answers are: x = 5, y = -1, and z = 2. 

compare to matrix solution:

```python
A = M[0:3,0:3]
A
```

```python
c = Matrix([ 10, 11, -6])
c
```

```python
solution = A.solve(c)
solution
```

both methods work!
